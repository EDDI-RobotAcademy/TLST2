```python
# csv 파일 DB로 저장하기

db = pymysql.connect(
    host = 'localhost',
    port = 3306,
    user = 'eddi',
    passwd = 'eddi@123',
    db = 'ztz_db'
)

f = open('product_data1.csv', 'r', encoding='UTF-8')
csv_reader = csv.reader(f, delimiter=',')

print(csv_reader)


cnt = 0

cursor = db.cursor()

for row in csv_reader:
    if cnt == 0:
        cnt += 1
        continue
    
    brand = (row[1])
    alcohol_type = (row[2])
    local = (row[3])
    favorite_num = (row[4])
    month_alcohol_check = (row[5])
    name = (row[6])
    price = (row[7])
    description = (row[8])
    product_images_name = (row[9])
    sub_title = (row[10])
    taste = (row[11])
    thumbnail_file_name = (row[12])
    view = (row[13])

        
    sql = "INSERT INTO product (brand,alcohol_type,local,favorite_num,month_alcohol_check,name,price,description,product_images_name,sub_title,taste,thumbnail_file_name,view) VALUES (%s, %s, %s, %s, %s, %s, %s,%s, %s, %s, %s, %s, %s)"
        
    cursor.execute(sql, (brand,alcohol_type,local,favorite_num,month_alcohol_check,name,price,description,product_images_name,sub_title,taste,thumbnail_file_name,view))

db.commit()

f.close()
```
